```python
# open text file for API key 
with open("api_key.txt") as file:
    API_key = file.read()
```

```python
#strip the text of the API to get rid of spaces and newline characters
API_key = API_key.strip()
```

```python
#import libraries
from alpha_vantage.timeseries import TimeSeries
import requests 
from bs4 import BeautifulSoup
import pandas as pd
import io
```

```python
# build TimeSeries variable from Alpha Vantage API
ts1 = TimeSeries(key = API_key)
```

```python
# get monthly stock data of Apple
ts1.get_monthly("AAPL")
```

```python
# get weekly stock data of Apple
ts1.get_weekly("AAPL")
```

```python
# get intraday stock data for Apple
ts1.get_intraday("AAPL")
```

```python
# get monthly stock of Apple with requests
url = 'https://www.alphavantage.co/query?function=TIME_SERIES_MONTHLY&symbol=IBM&apikey=' + str(API_key)
r = requests.get(url)
data = r.json()

print(data)
```

```python
# get weekly stock data 
url = url = 'https://www.alphavantage.co/query?function=TIME_SERIES_WEEKLY&symbol=IBM&apikey=' + str(API_key)
r = requests.get(url)
data = BeautifulSoup(r.content)
print(data)
```

```python
# get intraday stock data of Apple, valid interval values: 1, 5, 15, 30 and 60 minutes
url = 'https://www.alphavantage.co/query?function=TIME_SERIES_INTRADAY&symbol=IBM&interval=60min&apikey=' + str(API_key)
r = requests.get(url)
data = BeautifulSoup(r.content)

print(data)
```

```python
# get weekly stock data in DataFrame format
url = 'https://www.alphavantage.co/query?function=TIME_SERIES_WEEKLY&symbol=IBM&apikey=' + str(API_key) + '&datatype=csv'
r = requests.get(url).content
data = pd.read_csv(io.StringIO(r.decode("utf-8")))
print(data)
```
